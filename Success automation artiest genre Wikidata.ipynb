```python
import wptools
import pandas as pd
import numpy as np
import re

```

```python
df = pd.read_excel('name_test.xlsx')
df
```

```python
artists = df['Artist'].tolist()
#artists = ['Lady Gaga','AFI']
```

```python
artists
```

```python
pr_genres = []
for artist in artists:
############first level################
    try:
    
        page = wptools.page(artist)
        page.wanted_labels(['P136'])
        page.get_wikidata()
        data_pull = page.data['wikidata']
        genre = data_pull['genre (P136)']

        pr_genre = ''
        for s in genre:
            pr_genre = pr_genre + s + ','
            pr_genre = pr_genre[:-1]
        pr_genres.append(pr_genre)
     
 ################seconde level, plus band, and than again####################
    except:
        try:
            artist_new = artist[0]+ ' (band)'
            page = wptools.page(artist_new)
            page.wanted_labels(['P136'])
            page.get_wikidata()
            data_pull = page.data['wikidata']
            genre = data_pull['genre (P136)']
            
            pr_genre = ''
            for s in genre:
                pr_genre = pr_genre + s + ','
                pr_genre = pr_genre[:-1]
            pr_genres.append(pr_genre)
#########if level 1 and 2 both fail, give it n/a##########################        
        except:
        
            pr_genre = '**********Oops...This artist needs your manpower!'
            pr_genres.append(pr_genre)
            pass
        
```

```python
table = pd.DataFrame.from_dict({'artist_name': artists, 'genre':pr_genres})

seperate_table = table['genre'].str.replace('(',',').str.replace(')',',').str.split(",", n = 16, expand = True) 
seperate_table.insert(loc = 0, column = 'Artist_name', value = table['artist_name'])

column_lenth = seperate_table.shape[1] - 2

list_ = np.arange(1, 16, 1).tolist()
list_new = [x for x in list_ if x <=column_lenth ]
drop_colum = [i for i in list_new if i % 2 != 0  ]

seperate_table1 = seperate_table.drop(drop_colum,1)
seperate_table1.fillna(value=' ', inplace=True)
seperate_table1
```

```python
seperate_table1.to_excel('output3.xlsx')
```
